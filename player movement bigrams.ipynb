```python
import glob
import pandas as pd
import math
g = glob.glob('data/processed_games/[0-9]*.csv')

def get_strides(g):
    shifted = g.court_location.shift(1)
    shifted = shifted.apply(lambda x: (x,) if not pd.isnull(x) else (-1, ))
    orig = g.court_location.apply(lambda x: (x, ) if not pd.isnull(x) else (-1, ))
    g['movement_bigram'] = shifted+orig
    return g
    
total_counts = None
for n in g:
    print(n)
    df = pd.read_csv(n, low_memory=False)
    df = df[df.team_id != -1]
#     print(df.head())
    df_strides = df.groupby(['team_id', 'player_id', 'event_id']).apply(get_strides)
    df_strides.movement_bigram.apply(lambda x: (str(e) for e in x))
    df_counts = df_strides.groupby(['team_id', 'player_id', 'movement_bigram', 'of_def_flag'])['event_id'].count()
    if total_counts is None:
        total_counts = df_counts  
    else:
        total_counts = total_counts.add(df_counts, fill_value=0.0)
        print(len(total_counts.index))  
    
```

```python
total_counts.to_csv('player_movement_bigrams.csv')
```
